```python
#### plotting continued

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


x = np.linspace(0,30,500)
y = np.sin(x)*np.exp(-x/10)


fig,axes = plt.subplots(1,3,figsize = (9,3))

axes[0].plot(x,y,lw = 2)
axes[0].set_xlim(-5,35)
axes[0].set_ylim(-1,1)
axes[0].set_title('set_xlim/ylim')

axes[1].plot(x,y,lw = 2)
axes[1].axis('tight')
axes[1].set_title("axis('tight')")

axes[2].plot(x,y,lw = 2)
axes[2].axis('equal')
axes[2].set_title("axis('equal')")

```

```python
from scipy import linalg as la
from scipy import optimize 
import sympy 

x = np.linspace(-4,2,100)
y1 = (4-2*x)/3
y2 = (3-5*x)/4

plt.figure()
plt.plot(x,y1,'r',x,y2,'b',lw = 2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(['$2x_{1} + 3x_{2} - 4=0$','$5x_{1}+4x_{2}-3=0$'])
```

```python
A = sympy.Matrix([[2,3],[5,4]])
b = sympy.Matrix([4,3])
x = A.solve(b)
```

```python
### alternative
A = np.array([[2,3],[5,4]])
b = np.array([4,3])

x = la.solve(A,b)
```

```python
### rectangular A


x_var = sympy.symbols('x1,x2,x3')

A = sympy.Matrix([[1,2,3],[5,4,8]])
b = sympy.Matrix([7,8])
x = sympy.Matrix(x_var)

sympy.solve(A*x-b, x_var)
```
