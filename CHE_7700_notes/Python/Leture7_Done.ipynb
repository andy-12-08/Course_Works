```python
import csv


file = open('Exp_Data_V4_CSV.csv','r',encoding = 'utf-8-sig')
csvreader = csv.reader(file)

#header = []
header = next(csvreader)
#header2 = next(csvreader)
#sss = next(csvreader)

#header2
#header2[2]

rows = []
x = [] ### time
y1 = []
y2 = []
y3 = []
y4 = []

for row in csvreader:
    rows.append(row)
    x.append(float(row[0]))
    y1.append(float(row[1]))
    y2.append(float(row[2]))
    y3.append(float(row[3]))
    y4.append(float(row[4]))
type(x[0]) 
header
```

```python
myfile = open('first2.txt','w')
for i in rows:
    myfile.write(str(i)+'\n')
    
myfile.close()



with open('Out_CSV.csv','w',newline = '') as f:
    writer = csv.writer(f)
    writer.writerow(rows)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.figure(0)
plt.plot(x,y1,'r',x,y2,'b',x,y3,'k')
plt.title('Exp Plot',fontname = 'Times',fontsize = 20)
plt.xlabel('Time',fontname = 'Times',fontsize = 20)
plt.ylabel('Experiment',fontname = 'Times',fontsize = 20)
plt.legend(['Exp1','Exp2','Exp3'],prop = {'family':'Times','size':20})

plt.show()
```
